```python
import pandas as pd

df = pd.read_excel("API_Download_DS2_en_excel_v2_58494.xls", sheet_name = "Work Data")

df.head()
```

```python
df_melted = df.melt(
    id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
    var_name="Year",
    value_name="Value"
)

df_melted.head()
```

```python
important_indicators = [
    "GDP per capita",
    "Inflation rate",
    "Exchange rate",
    "Human Rights Development Index",
    "Unemployment rate",
    "Literacy rate",
    "Economic Freedom",
    "Interest rate",
    "Access to electricity",
    "Technicians in R&D",
    "Age dependency ratio, old (% of working-age population)",
    "Regulatory Quality: Percentile Rank",
    "Control of Corruption: Percentile Rank, Lower Bound of 90% Confidence Interval",
    "Foreign direct investment, net outflows",
    "Land area (sq. km)",
    "Population, total"
]

mask = df_melted["Indicator Name"].apply(
    lambda x: any(keyword.lower() in x.lower() for keyword in important_indicators)
)
df_filtered = df_melted[mask]
```

```python
df_pivoted = df_filtered.pivot_table(
    index=["Country Name", "Country Code", "Year"],
    columns="Indicator Name",
    values="Value"
).reset_index()

df_pivoted.columns.name = None

df_pivoted.head()
```

```python
print(df_pivoted.columns.tolist())
```

```python
df_pivoted = df_pivoted.drop(columns=["Access to electricity, rural (% of rural population)","Access to electricity, urban (% of urban population)","Age dependency ratio, old (% of working-age population)", "Government expenditure per student, primary (% of GDP per capita)", "Government expenditure per student, secondary (% of GDP per capita)", "Government expenditure per student, secondary (% of GDP per capita)","Government expenditure per student, tertiary (% of GDP per capita)","Literacy rate, adult female (% of females ages 15 and above)", "Literacy rate, adult male (% of males ages 15 and above)","Literacy rate, youth female (% of females ages 15-24)", "Literacy rate, youth male (% of males ages 15-24)", "Literacy rate, youth total (% of people ages 15-24)","Price level ratio of PPP conversion factor (GDP) to market exchange rate","Rural land area (sq. km)"])
df_pivoted.head()
```

```python
missing_percentage = df_pivoted.isna().mean() * 100
print(missing_percentage[missing_percentage > 0])
```

```python
df_pivoted = df_pivoted.drop(columns=["Literacy rate, adult total (% of people ages 15 and above)", "Literacy rate, youth (ages 15-24), gender parity index (GPI)", "Real effective exchange rate index (2010 = 100)", "Technicians in R&D (per million people)", "Urban land area (sq. km)"])
```

```python
columns_to_fill = ["Interest rate spread (lending rate minus deposit rate, %)", "Land area (sq. km)", "Lending interest rate (%)", "Real interest rate (%)"]
for col in columns_to_fill:
    df_pivoted[col] = df_pivoted[col].fillna(df_pivoted[col].mean())
```
